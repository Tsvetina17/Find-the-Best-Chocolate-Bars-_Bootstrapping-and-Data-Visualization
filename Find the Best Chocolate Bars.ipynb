# Finding the Best Chocolate Bars


## Executive Summary
The dataset for this project is provided by [DataCamp](app.datacamp.com) and I use it for an exercise in data science. The data covers diverse information about chocolate bars and their reviews. My aim is to find what are the characteristics of highly rated chocolate bars.

#### Data Cleaning and Validation
The dataframe doesn't contain any duplicates. There are 87 rows with missing values for 'num_ingredients' and 'ingredients'. There is no information why the data is missing. I examine the unique values in the columns and I determine that there are values in 'company_location' and 'bean_origin' that are entered in a wrong way so I make adjustments. Lastly, I split the data in 'ingredients' into a separate column of Boolean data type for each ingredient and create a new column representing the aggregated categories of 'rating'. I further validate the data during the exploratory data analysis.

#### Exploratory Data Analysis
There are 1605 bars produced by 580 manufactures in 65 locations over a period of 15 years and the beans come from 56 different countries. To explore the data I generate pair plots, estimate the correlation coefficients for the pairs and create additional visualizations. I make the observations that most of the chocolate bars have two or three ingredients. The most common bean origin is from Venezuela. The majority of companies is in the U.S.A. and Soma appears to be a top manufacturer. The most frequent combination of ingredients is bean, sugar, and cocoa butter.

Furthermore, the distribution of 'rating' is left-skewed with a few outliers. The normality test shows that there is not enough evidence suggesting that the data for 'rating' has normal distribution. Most of the chocolate bars have rating of 3 or higher. The mean rating is on the rise during the period and the variability is decreasing. The number of reviews on average increased until 2015 but there was a decrease afterwards. There were a few sharp declines in 2017 and 2020. Some ratings of 'Unpleasant' were reported in 2006 - 2012 but there were no such negative ratings afterwards. The percent of cocoa in chocolate bars is on average 70% to 73% over the period and the variability has been decreasing. 

#### Average rating by origin of the beans
Each manufacturer has between 1 and 22 bean origins with more than 75% having only up to 4 bean origins. There are a lot of bean origins with a very low count of rating so the data isn't very robust. Thus, I focus on countries with more than five ratings. Solomon Islands has the highest mean rating of 3.45 but the count is 10 which is not very robust. The other Top 20 countries have mean rating between 3.18 and 3.29 which demonstrates little variability. However, the spread of the values is very different. It is wider and with outliers for countries with higher count of rating such as Venezuela, Dominican Republic and Peru. Almost all countries in the dataframe have a median of 3 or 3.25. Overall, bean origin doesn't seem to be an indicator of quality.

#### Average rating by company location
There are a lot of values in 'company_location' that have a count of 'rating' less than 5. This can create a potential problem since company locations that have very few values for 'rating' can be sensitive to extreme values. I remove them to minimize noise in the data. Poland has the highest mean rating of 3.38 but it has only eight reviews so the data for it is not very robust. The rest of the countries in Top 20 all have mean rating between 3.21 and 3.36. Hence, there is little variability in the values for mean rating in these company locations. The spread is wider and with outliers for countries with higher count of rating such as France, Canada and Switzerland. France, Italy and Germany are the only countries that have an outlier with score of 1.5. The top three countries have the same third quartile value of 3.75. To sum up, there is little variety in mean ratings and the limited count of ratings makes the data less robust for some company locations. The company location doesn't seem to greatly affect the rating.

#### Average rating by manufacturer
I use data only for manufacturers that have count of rating over 10 in order to minimize noise in the data. The data shows a lot of variability. Idilio (Felchlin) from Switzerland has the highest mean rating of 3.78 while Letterpress from the U.S.A. is number 20 with a mean of 3.41. One of the Top 20 manufacturers has an outlier of 1.5 and all but one have values of 4.0. It seems that 'manufacturer' has a lot more variability than 'company_location' or 'bean_origin'.

#### Cocoa content and rating of chocolate bars
Chocolate bars in rating group 'Unpleasant' have 85% mean cocoa content compared to 71% for 'Outstanding'. When I plot data and regression model for 'cocoa_percent' and 'rating_categories' it becomes evident that bars with rating 'Outstanding' have the lowest variability of cocoa content. This is the only category that doesn't have any bars with cocoa content over 90%. The regression model shows that there is a weak correlation between the variables, i.e. there is a slight decline in average cocoa content for bars with higher rating.

#### Characteristics of chocolate bars with rating over 3.5
60% of the chocolate bars with rating over 3.5 have 'cocoa_percent' of 70-72% and 76% of them have two or three ingredients. Only 8% have five ingredients. There are some differences when the chocolate bars were reviewed with a peak in 2016 and dips in 2008 and 2020. 'B, S, C' and 'B, S' are the two top combinations of ingredients in terms of mean rating and they also have the highest count. There are quite a lot of combinations that have count under 10 so there seems to be a lot of noise in the data. I exclude them and it becomes more obvious that a small amount of the bars contains sweeteners and they tend to have lower mean of rating. Salt is found in only 24 bars that have a mean rating of 3.25.

#### Comparison of reviews for highly rated chocolate bars and other bars
'nutty' is among the top words and equally common in reviews for both groups. 'spicy' and 'mild' are also used fairly often. Chocolate bars with rating > 3.5 are slightly more often called 'creamy' and 'rich'. Their reviews are more likely to contain words describing specific flavors such as 'caramel', 'strawberry', 'cherry', 'brownie'. In contrast, chocolate bars with rating <= 3.5 are often 'earthy', 'sandy', 'fatty', 'sour', 'sweet', 'bitter' and 'vanilla'.

#### Average rating of chocolate bars with and without lecithin
The mean of the rating is 3% higher for chocolate bars with lecithin (3.23) compared to those without lecithin (3.14) but I need to check if this difference is statistically significant. A normality test shows that the sample data hasn't been drawn from a normally distributed population. Therefore, I use bootstrapping which is a method that uses random sampling with replacement to estimate a population parameter. The result suggests significant empirical evidence that the mean rating of chocolate bars without lecithin is higher than that of bars with lecithin.

```python
# Adjust the values in 'bean_origin' that were entered in the wrong way
df['bean_origin'] = df['bean_origin'].str.replace('DR Congo', 'Congo')
df['bean_origin'] = df['bean_origin'].str.replace('Sao Tome & Principe', 'Sao Tome')
df['bean_origin'] = df['bean_origin'].str.replace('Principe', 'Sao Tome')
df['bean_origin'] = df['bean_origin'].str.replace('Sulawesi', 'Indonesia')
df['bean_origin'] = df['bean_origin'].str.replace('Sumatra', 'Indonesia')
df['bean_origin'] = df['bean_origin'].str.replace('Tobago', 'Trinidad')
```

```python
# Print out the unique values in 'ingredients'
df['ingredients'].unique()
```

According to the description of the data, 'ingredients' contains: B (Beans), S (Sugar), S* (Sweetener other than sugar or beet sugar), C (Cocoa Butter), (V) Vanilla, (L) Lecithin, (Sa) Salt. I split the data in 'ingredients' and make a separate column for each ingredient of bool type

```python
# Create a column of boolean type for each ingredient 
df['beans'] = df['ingredients'].str.contains('B')
df['sugar'] = df['ingredients'].str.contains('S')
df['sweetener'] = df['ingredients'].str.contains('S\*')
df['cocoa_butter'] = df['ingredients'].str.contains('C')
df['vanilla'] = df['ingredients'].str.contains('V')
df['lecithin'] = df['ingredients'].str.contains('L')
df['salt'] = df['ingredients'].str.contains('Sa')
```

According to the description of the data, the column 'rating' contains data grouped in the following way:
 - 1.0-1.9 Unpleasant
 - 2.0-2.9 Disappointing
 - 3.0-3.49 Recommended
 - 3.5-3.9 Highly Recommended
 - 4.0-5.0 Oustanding

I create an additional column in the dataframe to represent these ratings.

```python
# Create a new column representing the groups of 'rating'
df['rating_groups'] = df['rating'].apply(lambda x: 'Unpleasant' if x>=1 and x<2
                                                else 'Disappointing' if x>=2 and x<3
                                                else 'Recommended' if x>=3 and x<3.5
                                                else 'Highly recommended' if x>=3.5 and x<4
                                                else 'Outstanding')
```

```python
# Create a new column representing the categories of 'rating' but with values of 1-5 this time
df['rating_categories'] = df['rating'].apply(lambda x: 0 if x>=1 and x<2
                                                else 1 if x>=2 and x<3
                                                else 2 if x>=3 and x<3.5
                                                else 3 if x>=3.5 and x<4
                                                else 4)
```

The distribution of 'rating' is left-skewed with a few outliers. Most of the chocolate bars have rating of 3 or higher. I conduct a test to confirm that 'rating' doesn't have a normal distribution. H0: Sample comes from a normal distribution; H1: Sample doesn't come from a normal distribution.

```python
ratings = df['rating'].to_numpy()
k2, p = stats.normaltest(ratings)
alpha = 5e-2

if p < alpha:
    print('Reject null hypothesis at alpha =', alpha)

else:
    print('Cannot reject null hypothesis at alpha =', alpha)

```

The normality test shows that there is not enough evidence suggesting that the data for 'rating' has normal distribution.

## Average rating of chocolate bars with and without lecithin

```python
# Make a copy of the original dataframe and drop lines with no information about ingredients
df_copy = df.copy()
df_copy.dropna(subset = 'num_ingredients', inplace = True)
```

```python
# Compare the means of the rating of the two groups of chocolate bars (with or without lecithin)
df_lecithin = df_copy[df_copy['lecithin'] == True]
df_without_lecithin = df_copy[df_copy['lecithin'] == False]
avg_rating_with_L = df_lecithin[df_lecithin['lecithin'] == True]['rating'].mean()
avg_rating_without_L = df_without_lecithin[df_without_lecithin['lecithin'] == False]['rating'].mean()
print('The average rating for chocolate bars with lecithin is', round(avg_rating_with_L,2), 'and for bars without lecithin is', round(avg_rating_without_L,2))
```

The mean of the rating is 3% higher for chocolate bars with lecithin (3.23) compared to those without lecithin (3.14). In order to check if this difference is statistically significant I need to conduct additional tests. I start by checking if the distribution of 'rating' for bars with or without lecithin is normal.

```python
# Plot the distribution of 'rating' for chocolate bars with and without lecithin
plt.subplot(1, 2, 1)
sns.histplot(x = df_lecithin['rating'])
plt.title("'rating' of bars without lecithin")

plt.subplot(1, 2, 2)
sns.histplot(x = df_without_lecithin['rating'])
plt.title("'rating' of bars with lecithin")
```

The distributions of 'rating' for bars with and without lecithin doesn't visually appear to be normal distribution. I use a normality test to determine whether the sample data has been drawn from a normally distributed population. normaltest from scipy.stats tests the null hypothesis that a sample comes from a normal distribution. It is based on D’Agostino and Pearson’s test.

```python
# Normality test of 'rating' for bars with lecithin
rating_lecithin = df_lecithin['rating'].to_numpy()

k2, p = stats.normaltest(rating_lecithin)
alpha = 0.05

if p < alpha:
    print('Reject null hypothesis at alpha =', alpha)
else:
    print('Cannot reject null hypothesis at alpha =', alpha)
```

```python
# Normality test of 'rating' for bars without lecithin
rating_without_lecithin = df_without_lecithin['rating'].to_numpy()

k2, p = stats.normaltest(rating_without_lecithin)
alpha = 0.05

if p < alpha:
    print('Reject null hypothesis at alpha =', alpha)
else:
    print('Cannot reject null hypothesis at alpha =', alpha)
```

Since the distribution of the variables is not normal, I will use bootstrapping. This is a method that uses random sampling with replacement to estimate a population parameter. I learned how to write the following code for conducting bootstrapping in the course 'Statistical Thinking in Python Part 2'.

```python
# Set a random seed
np.random.seed(42)

# Extract ratings for bars with and without lecithin
ratings_with_lecithin = df.loc[df['lecithin'] == True]['rating'].to_numpy()
ratings_without_lecithin = df.loc[df['lecithin'] == False]['rating'].to_numpy()

# Number of simulations
n_simulations = 10000

# Compute the difference of the means in the samples
mean_diff = round(np.mean(rating_without_lecithin) - np.mean(ratings_with_lecithin), 2)

# Helper functions for bootstrapping
def bootstrap_replicate_1d(data, func):
    '''Generate bootstrap replicate of 1D data'''
    bs_sample = np.random.choice(data, len(data))

def draw_bs_reps(data, func, size = 1):
    '''Draw bootstrap replicates.'''

    # Initialize array of replicates: bs_replicates
    bs_replicates = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_replicates[i] = bootstrap_replicate_1d(data, func)

    return bs_replicates

# Bootstrap replicates of means
bootstrap_without_lecithin = draw_bs_reps(ratings_without_lecithin, np.mean, n_simulations)
bootstrap_with_lecithin = draw_bs_reps(ratings_with_lecithin, np.mean, n_simulations)
```

```python
# Compute samples of difference of means: bs_diff_replicates
bs_diff_replicates = np.round(bootstrap_without_lecithin - bootstrap_with_lecithin, 2)

# Compute 95% confidence interval: conf_int
conf_int = np.percentile(bs_diff_replicates, [2.5, 97.5])

print('The difference of means is', mean_diff, 'rating points')
print('95% confidence interval = ', conf_int, 'rating points')
```

```python
np.random.seed(42)

# Compute mean of combined data sets
combined_mean = np.mean(np.concatenate((ratings_with_lecithin, ratings_without_lecithin)))

# Shift the samples
no_lecithin_shifted  = ratings_without_lecithin - np.mean(rating_without_lecithin) + combined_mean
lecithin_shifted = ratings_with_lecithin - np.mean(ratings_with_lecithin) + combined_mean

# Get bootstrap replicates of shifted data sets
bs_replicates_no_lecithin_shifted = draw_bs_reps(no_lecithin_shifted, np.mean, n_simulations)
bs_replicates_lecithin_shifted = draw_bs_reps(lecithin_shifted, np.mean, n_simulations)

# Compute replicates of difference of means
bs_diff_replicates = bs_replicates_no_lecithin_shifted - bs_replicates_lecithin_shifted

# Compute the p value
p = np.sum(bs_diff_replicates >=  mean_diff/len(bs_diff_replicates))

print(round(p,4))
```

The resulting p value of the analysis with bootstrapping is less than 0.05 which suggests significant empirical evidence that the mean rating of chocolate bars without lecithin is higher than that of bars with lecithin.
